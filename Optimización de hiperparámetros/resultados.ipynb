```python
import matplotlib.pyplot as plt
import re

# Función para leer y mostrar el contenido del archivo
def leer_mostrar_archivo(filename):
    try:
        with open(filename, 'r') as file:
            data = file.read()
            print("Contenido del archivo:")
            print(data)
            return data
    except FileNotFoundError:
        print("El archivo no fue encontrado.")
        return None

# Función para mostrar una imagen
def mostrar_imagen(image_path):
    try:
        img = plt.imread(image_path)
        plt.imshow(img)
        plt.axis('off')  # No mostrar los ejes
        plt.show()
    except FileNotFoundError:
        print("La imagen no fue encontrada.")

# Función para extraer y graficar la lista 'history_best_cost_per_iteration' desde el texto del archivo
def extraer_y_graficar_history(data):
    if data:
        match = re.search(r"History best cost per iteration: \[([-\d.,\s]+)\]", data)
        if match:
            history_list = match.group(1).split(',')
            history_floats = [float(num.strip()) for num in history_list]
            plt.figure(figsize=(10, 5))
            plt.plot(history_floats, marker='o', linestyle='-')
            plt.title('History Best Cost per generation')
            plt.xlabel('Generation')
            plt.ylabel('Cost')
            plt.grid(True)
            plt.show()
        else:
            print("No se encontró la lista 'History best cost per iteration' en el archivo.")


hyperparameters = {
        'rec_hidden_layers_config': "layer_configurations",
        'activation': ['relu', 'tanh', 'sigmoid', 'leaky_relu', 'elu'],
        'regularization': ['none', 'dropout_0.1', 'dropout_0.2', 'dropout_0.5'],
        'optimizer': ['sgd', 'momentum', 'adam', 'rmsprop', 'adamw'],
        'learning_rate': [0.1, 0.01, 0.001, 0.0005, 0.0001, 0.00005, 0.00001],
        'epochs': [10, 20, 50, 100, 200],
        'batch_size': [16, 32, 64, 128, 256, 512]
    }
hyperparameter_names = list(hyperparameters.keys())

import EDAspy2
from EDAspy2 import visualization as Vis
import pickle
```

```python
# Nombres de archivo
filename_text = 'resultado_ebna1.txt'
filename_image = 'tresuno.png'

# Ejecución de funciones
contenido = leer_mostrar_archivo(filename_text)
mostrar_imagen(filename_image)
extraer_y_graficar_history(contenido)
```
