```python
import json
import matplotlib.pyplot as plt
import numpy as np

def leer_datos(filename):
    with open(filename, 'r') as file:
        return [json.loads(line) for line in file]

def graficar_ntk(data, gen_size):
    ntk_norms = [d['ntk_nuclear_norm'] for d in data]
    ntk_diffs = [d['ntk_difference'] for d in data]
    
    fig, ax = plt.subplots(2, 1, figsize=(14, 10), sharex=True)
    ax[0].plot(ntk_norms, label='NTK Nuclear Norm', marker='o', linestyle='-')
    ax[1].plot(ntk_diffs, label='NTK Difference', marker='o', linestyle='-')
    
    for i in range(gen_size, len(data), gen_size):
        ax[0].axvline(x=i, color='r', linestyle='--')
        ax[1].axvline(x=i, color='r', linestyle='--')
    
    ax[0].set_title('NTK Nuclear Norm over Generations')
    ax[1].set_title('NTK Difference over Generations')
    ax[0].set_ylabel('Norm')
    ax[1].set_ylabel('Difference')
    ax[1].set_xlabel('Individual Index')
    
    ax[0].legend()
    ax[1].legend()
    ax[0].grid(True)
    ax[1].grid(True)
    plt.tight_layout()
    plt.show()

def graficar_metrics(data, metric, gen_size):
    fig, ax = plt.subplots(figsize=(14, 6))
    
    max_epochs = max(len(individual[f'train_{metric}']) for individual in data)
    offset = 0
    
    for i, individual in enumerate(data):
        num_epochs = len(individual[f'train_{metric}'])
        x = np.array(range(1, num_epochs + 1)) + offset
        ax.plot(x, individual[f'train_{metric}'], label='Train' if i == 0 else "", marker='o', linestyle='-', color='tab:blue')
        ax.plot(x, individual[f'validation_{metric}'], label='Validation' if i == 0 else "", marker='o', linestyle='-', color='tab:orange')
        
        offset += num_epochs + 0.5
    
    generation_offset = 0
    for gen in range(gen_size, len(data), gen_size):
        generation_offset += sum(len(data[j][f'train_{metric}']) for j in range(gen - gen_size, gen)) + 3
        ax.axvline(x=generation_offset, color='r', linestyle='--', linewidth=1)

    ax.set_title(f'{metric.title()} over Generations')
    ax.set_xlabel('Individual Index and Epochs')
    ax.set_ylabel(metric.title())
    ax.legend(loc='upper right')
    ax.grid(True)
    plt.tight_layout()
    plt.show()
```

```python
# Leer datos
filename = 'individual_info1.txt'
data = leer_datos(filename)

# Graficar NTK
graficar_ntk(data, 6)

# Graficar loss y accuracy
graficar_metrics(data, 'loss', 6)
graficar_metrics(data, 'accuracy', 6)
```
